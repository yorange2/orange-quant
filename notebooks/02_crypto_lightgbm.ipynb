# 🪙 加密货币量化实验 — LightGBM

数据来源：Hyperliquid Perpetual Futures
币种：BTC, ETH, XRP, BNB, SOL

```python
import qlib
from qlib.data import D

# 初始化 crypto 数据
qlib.init(provider_uri="../data/qlib_data/crypto", region="cn")
```

```python
# 快速看一下 BTC 的价格走势
from qlib.data.dataset.loader import QlibDataLoader

loader = QlibDataLoader(config={
    "feature": ["$close"],
    "label": ["Ref($close, -1) / $close - 1"],  # 次日收益率
})
df = loader.load("all", start_time="2023-06-01", end_time="2026-06-29")
print(df.head(10))
print(f"\n总样本数: {len(df)}")
```

```python
# ── 运行 LightGBM 实验 ──
from orange_quant.workflow.experiment import run_from_yaml

results = run_from_yaml("../config/workflow_config_crypto.yaml")

print(f"\n预测样本数: {len(results['predictions'])}")
print(f"预测前10条:\n{results['predictions'].head(10)}")
```

```python
# ── 结果分析 ──
recorder = results["recorder"]

print("=" * 50)
print("模型参数")
print("=" * 50)
for k, v in recorder.list_params().items():
    if k != "cmd-sys.argv":
        print(f"  {k}: {v}")

print("\n" + "=" * 50)
print("信号分析 (IC)")
print("=" * 50)
for k, v in recorder.list_metrics().items():
    if "IC" in k or "Rank" in k:
        print(f"  {k}: {v:.4f}")

print("\n" + "=" * 50)
print("回测绩效")
print("=" * 50)
for k, v in recorder.list_metrics().items():
    if "1day" in k:
        print(f"  {k}: {v:.4f}" if isinstance(v, float) else f"  {k}: {v}")
```

## 注意事项

- 加密货币 24/7 交易，qlib 的 "day" 频率实际上是 Hyperliquid 的日线（UTC 0点对齐）
- 只有 5 个币种，`topk=3, n_drop=1` 的配置比较合适
- 数据从 2023-06 开始，历史较短，模型可能不够稳定
- 如需更多币种，修改 `orange_quant/data/crypto_download.py` 中的 `DEFAULT_COINS`
